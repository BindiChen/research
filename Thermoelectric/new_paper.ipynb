```python
import pandas as pd
import numpy as np
from algorithm import get_total_mass
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
```

```python
def load_html_table(file):
    with open(file, 'r') as f:
        dfs = pd.read_html(f.read())
    return dfs
```

## Concat all tables

```python
allData = pd.concat([df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8, df_9], ignore_index=True)
allData
```

## Cleaning

```python
df_plot = allData[pd.to_numeric(allData['zT'], errors='coerce').notnull()]
df_plot = df_plot[df_plot['total_mass']>0]
```

```python
df_plot.info()
```

```python
df_plot['zT'] = df_plot['zT'].astype('float')
```

```python
df_plot = df_plot.reset_index()
df_plot
```

```python
df_plot.info()
```

```python
# Figure
fig, ax = plt.subplots()
fig.set_size_inches(16, 10)

#labels
ax.set_title('Relationship between composition mass and zT', fontsize=20)
ax.set_xlabel('Composition Mass', fontsize=18)
ax.set_ylabel('zT', fontsize=18)


plt.scatter(df_plot['total_mass'], df_plot['zT'], c=df_plot['zT'], s=250, cmap='RdYlGn', alpha=0.7)
cbar = plt.colorbar()
cbar.set_label('Color Intensity', fontsize=18)
plt.show()
```
