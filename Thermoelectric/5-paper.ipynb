## Helper functions

```python
def load_html_table(file):
    with open(file, 'r') as f:
        dfs = pd.read_html(f.read())
    return dfs
```

# Concat all tables

```python
allData = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8], ignore_index=True)
allData
```

## Cleaning
* removing zT with non-numeric value
* removing total_mass=NaN

```python
df_plot = allData[pd.to_numeric(allData['zT'], errors='coerce').notnull()]
df_plot = df_plot[df_plot['total_mass']>0]
```

```python
df_plot = df_plot.reset_index()
df_plot
```

### Plotting

```python
# Figure
fig, ax = plt.subplots()
fig.set_size_inches(16, 10)

#labels
ax.set_title('Relationship between composition mass and zT', fontsize=20)
ax.set_xlabel('Composition Mass', fontsize=18)
ax.set_ylabel('zT', fontsize=18)


plt.scatter(df_plot['total_mass'], df_plot['zT'], c=df_plot['zT'], s=250, cmap='RdYlGn', alpha=0.7)
cbar = plt.colorbar()
cbar.set_label('Color Intensity', fontsize=18)
plt.show()
```

```python
## Removing outlier - data from Table 6, 7 and 8
df_plot_2 = df_plot.iloc[:-11,:]
```

```python
# Figure
fig, ax = plt.subplots()
fig.set_size_inches(16, 10)

#labels
ax.set_title('Relationship between composition mass and zT', fontsize=20)
ax.set_xlabel('Composition Mass', fontsize=18)
ax.set_ylabel('zT', fontsize=18)


plt.scatter(df_plot_2['total_mass'], df_plot_2['zT'], c=df_plot_2['zT'], s=250, cmap='RdYlGn', alpha=0.7)
cbar = plt.colorbar()
cbar.set_label('Color Intensity', fontsize=18)
plt.show()
```
